## ¿Qué es Python?

En términos sencillos, Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al inglés. Es famoso por ser "legible": leer código en Python es casi como leer instrucciones en un manual.

Es un lenguaje **versátil** (sirve para todo, desde apps web hasta inteligencia artificial) y **comunitario** (hay miles de librerías gratuitas listas para usar).

## 3 Ejemplos Prácticos en Python

Aquí tienes tres scripts básicos que ilustran cómo se aplica la lógica de programación en tu campo:

### 1. Cálculo de Densidad y Control de Calidad

Este código ayuda a determinar si un lote de producto líquido (como leche o jugo) cumple con un estándar de densidad específico.

```
# Cálculo de densidad simple
masa = float(input("Ingresa la masa de la muestra (g): "))
volumen = float(input("Ingresa el volumen de la muestra (ml): "))

densidad = masa / volumen

print(f"La densidad calculada es: {densidad:.2f} g/ml")

if 1.028 <= densidad <= 1.033:
    print("Estado: La densidad es óptima para leche entera.")
else:
    print("Estado: Fuera de rango. Revisar posible adulteración.")
```

### 2. Estimación de Pérdida de Peso por Secado

En la agroindustria, el secado es clave. Este código calcula cuánta agua se eliminó de un producto.

```
# Balance de masa en secado
peso_inicial = 500.0  # kg
humedad_inicial = 0.20 # 20%
humedad_final = 0.05   # 5%

# Cálculo de sólidos totales (que permanecen constantes)
solidos = peso_inicial * (1 - humedad_inicial)

# Peso final del producto
peso_final = solidos / (1 - humedad_final)

agua_evaporada = peso_inicial - peso_final

print(f"Peso final del producto: {peso_final:.2f} kg")
print(f"Agua total evaporada: {agua_evaporada:.2f} kg")
```

### 3. Monitor de Temperatura para Almacenamiento

Simulación de un sistema que alerta si un cuarto frío supera la temperatura crítica.

```
# Simulador de sensor de temperatura
temperaturas = [3.2, 4.1, 4.5, 5.2, 4.8, 6.1, 3.9] # Datos de 7 horas
limite_maximo = 5.0

print("Análisis de cadena de frío:")
for t in temperaturas:
    if t > limite_maximo:
        print(f"ALERTA: Temperatura de {t}°C excede el límite!")
    else:
        print(f"Temperatura estable: {t}°C")
```
