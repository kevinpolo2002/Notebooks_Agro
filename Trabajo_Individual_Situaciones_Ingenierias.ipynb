**Ejercicio 1**: Capacidad de Silos de Granos

1. **Problema**: Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico. El usuario ingresa el radio y la altura del silo en metros.
2. **Datos**: Densidad del maíz ≈0.75 ton/m3
. Volumen V=π×r2×h
3. **Prueba**: Radio = 3m, Altura = 10m. Salida ≈212.06
 toneladas.

```python
import math

# Entrada de datos
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

# Proceso
# Calculamos el volumen: V = pi * r^2 * h
volumen = math.pi * (radio ** 2) * altura
# Calculamos las toneladas: Ton = Volumen * Densidad (0.75)
toneladas = volumen * 0.75

# Salida
print(f"La capacidad del silo es: {toneladas:.2f} toneladas")
```

# **Explicación del Código Paso a Paso**

1. import math: Esta línea trae a nuestro programa la librería de matemáticas de Python, que contiene el valor exacto de $\pi$ (math.pi)
2. radio = float(input(...)): Solicita al usuario el radio, lo lee como texto y lo convierte a un número decimal (float) para poder hacer cálculos.
3. altura = float(input(...)): Realiza la misma acción que la línea anterior, pero para obtener el dato de la altura del silo.
4. volumen = math.pi * (radio ** 2) * altura: Aplica la fórmula geométrica. El símbolo ** 2 eleva el radio al cuadrado. El resultado se guarda en la variable volumen.
5. toneladas = volumen * 0.75: Multiplica el espacio total del silo por la densidad del maíz para saber cuánto pesa el grano almacenado.
6. print(...): Muestra el resultado final en la pantalla. El código :.2f sirve para que solo se muestren dos decimales, haciendo que el resultado se vea más limpio.

**Ejercicio 2**: Dosificación de Fertilizante Líquido

1. **Problema**: Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea. El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).
2. **Datos**: El fertilizante se diluye en agua en una relación 1:500 (1 litro de producto por 500 de agua).
3. **Prueba**: 5 hectáreas, Dosis 2 L/ha. Salida: 10L de producto y 5000L de agua total.

```python
# Entrada de datos
hectareas = float(input("Ingrese el número de hectáreas (e.g., 5): "))
dosis_por_ha = float(input("Ingrese la dosis por hectárea en L/ha (e.g., 2): "))

# Proceso
producto_total = hectareas * dosis_por_ha
agua_total = producto_total * 500

# Salida
print(f"Producto necesario: {producto_total} L")
print(f"Agua necesaria para dilución: {agua_total} L")
```

# **Explicación del código**

 + hectareas = float(input(...)): Esta línea muestra un mensaje en pantalla pidiendo la extensión del terreno. La función input() recibe el texto y float() lo convierte en un número con decimales para poder realizar cálculos exactos.
+ dosis_por_ha = float(input(...)): Funciona igual que la anterior, capturando cuántos litros de fertilizante se necesitan por cada hectárea individual.
+ producto_total = hectareas * dosis_por_ha: Aquí realizamos la primera operación lógica. Multiplicamos el tamaño del terreno por la dosis unitaria para obtener la cantidad total de fertilizante puro (el "producto") que se requiere.
+ agua_total = producto_total * 500: Basándonos en la relación de dilución $1:500$, multiplicamos los litros de producto por $500$ para calcular cuánta agua se necesita para diluirlo correctamente.
+ print(...): Finalmente, usamos estas líneas para mostrar al usuario los dos resultados calculados, permitiendo que sepa exactamente cuánto preparar de cada componente.

**Ejercicio 3**: Rendimiento de Pulpa en Frutas

1. **Problema**: Una planta procesadora necesita saber el porcentaje de aprovechamiento. Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).
2. **Datos**: Pulpa=Peso Total−Residuos
. %Rendimiento=(Pulpa/Peso Total)×100
3. **Prueba**: Peso Total = 500kg, Residuos = 150kg. Salida: 350kg de pulpa (70%).

```python
# Entrada de datos
peso_total = float(input("500 (kg): "))
residuos = float(input("150 (kg): "))

# Proceso
pulpa = peso_total - residuos
porcentaje_rendimiento = (pulpa / peso_total) * 100

# Salida
print(f"Cantidad de pulpa obtenida: {pulpa} kg")
print(f"Porcentaje de rendimiento: {porcentaje_rendimiento}%")
```

# **Explicación del código**

+ peso_total = float(input(...)): Solicita al usuario la cantidad total de fruta que ingresa a la planta en kilogramos. Se usa float para permitir valores precisos (ej. 500.5 kg).
+ residuos = float(input(...)): Captura el peso de las partes que no se aprovechan, como cáscaras y semillas.
+ pulpa = peso_total - residuos: Realiza la resta lógica para determinar cuánta materia prima útil queda después de limpiar la fruta.
+ porcentaje_rendimiento = (pulpa / peso_total) * 100: Calcula la eficiencia del proceso dividiendo la parte útil entre el total original y convirtiéndolo a porcentaje.
+ print(...): Muestra al usuario ambos resultados: los kilogramos netos de pulpa y qué tan eficiente fue el aprovechamiento en términos porcentuales.
